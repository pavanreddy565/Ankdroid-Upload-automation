```python
df = pd.read_excel('/content/Infosys Vocabulary.xlsx')
df.head()
```

```python

import genanki
```

```python
import random
model_id =random.randrange(1 << 30, 1 << 31)
```

```python
css='''
  .card {
    font-family: arial;
    font-size: 20px;
    text-align: left;
    color: black;
    background-color: white;
  }
  h1 {
    color: #2e6c80;
  }
  p {
    margin: 0.3em 0;
  }
'''
```

```python
my_model = genanki.Model(
    model_id,
    'Simple Model',
    fields=[
        {'name': 'Word'},
        {'name': 'Answer'},
        {'name': 'Example'}
    ],
    templates=[
        {
            'name': 'Card 1',
            'qfmt': '<h1>{{Word}}</h1>',
            'afmt': '''
                {{FrontSide}}
                <hr id="answer">
                <p><b>Meaning:</b> {{Answer}}</p>
                <p><b>Example:</b> {{Example}}</p>
            ''',
        },
    ]
)
```

```python
deck_id = random.randrange(1 << 30, 1 << 31)
```

```python
my_deck = genanki.Deck(
  deck_id,
  'Infosys Vocabulary')
```

```python
for _, row in df.iterrows():
    fields = [
        str(row['word']) if pd.notna(row['word']) else '',
        str(row['Meaning']) if pd.notna(row['Meaning']) else '',
        str(row['Example Sentence']) if pd.notna(row['Example Sentence']) else ''
    ]

    note = genanki.Note(
        model=my_model,
        fields=fields
    )
    my_deck.add_note(note)
```

```python
genanki.Package(my_deck).write_to_file('output.apkg')
```
